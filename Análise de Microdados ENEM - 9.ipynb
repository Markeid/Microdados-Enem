```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importando Atlas
atlasHumano = pd.read_csv("C:/Users/BlueShift/Desktop/brutos/atlas2013_dadosbrutos.csv", 
sep=",")
```

```python
MediaEnemIDHM = pd.merge(mediaEnemporMunicipio, atlasHumano, how="left", left_on=['CO_MUNICIPIO_ESC'], right_on=['Codmun7'])
```

```python
MediaEnemIDHM.head()
```

```python
MediaEnemIDHM = MediaEnemIDHM.filter(items=['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC',	'CO_UF_ESC', 'SG_UF_ESC', 'NU_NOTA_CN'
                                            'NU_NOTA_CH', 'NU_NOTA_LC',	'NU_NOTA_MT', 'NU_NOTA_REDACAO',
                                            'IDHM',	'IDHM_E', 'IDHM_L',	'IDHM_R'])
```

```python
MediaEnemIDHM
```

```python
MediaEnemIDHM.plot.scatter('NU_NOTA_REDACAO', 'IDHM', figsize=(10, 10))
```

```python
MediaEnemIDHM.plot.scatter('NU_NOTA_MT', 'IDHM', figsize=(10, 10))
```

```python
MediaEnemIDHM.dtypes
```

```python
mean_REDACAO= MediaEnemIDHM.NU_NOTA_REDACAO.mean()
mean_IDHM= MediaEnemIDHM.IDHM.mean()
```

```python
mean_REDACAO
```

```python
mean_IDHM
```

```python
idhmMunicipio = MediaEnemIDHM.IDHM
notaMunicipio = MediaEnemIDHM.NU_NOTA_REDACAO

q1 = (idhmMunicipio < mean_IDHM) & (notaMunicipio < mean_REDACAO)
q2 = (idhmMunicipio > mean_IDHM) & (notaMunicipio < mean_REDACAO)
q3 = (idhmMunicipio > mean_IDHM) & (notaMunicipio > mean_REDACAO)
q4 = (idhmMunicipio < mean_IDHM) & (notaMunicipio > mean_REDACAO)

fig, ax = plt.subplots(2,2, figsize=(15,7), sharex='col', sharey='row')
plt.subplots_adjust(wspace=0, hspace=0)

minRedacao =MediaEnemIDHM.NU_NOTA_REDACAO.min()
maxRedacao =MediaEnemIDHM.NU_NOTA_REDACAO.max()
minIDHM =MediaEnemIDHM.IDHM.min()
maxIDHM =MediaEnemIDHM.IDHM.max()

MediaEnemIDHM1 = MediaEnemIDHM[q1]
MediaEnemIDHM1.plot.scatter('NU_NOTA_REDACAO', 'IDHM', s=6, ax = ax[1,0], xlim = (minRedacao, mean_REDACAO), ylim =(minIDHM, mean_IDHM))

MediaEnemIDHM2 = MediaEnemIDHM[q2]
MediaEnemIDHM2.plot.scatter('NU_NOTA_REDACAO', 'IDHM', s=6, ax = ax[0,0], xlim = (minRedacao, mean_REDACAO), ylim =(maxIDHM, mean_IDHM))

MediaEnemIDHM3 = MediaEnemIDHM[q3]
MediaEnemIDHM3.plot.scatter('NU_NOTA_REDACAO', 'IDHM', s=6, ax = ax[0,1], xlim = (maxRedacao, mean_REDACAO), ylim =(maxIDHM, mean_IDHM))

MediaEnemIDHM4 = MediaEnemIDHM[q4]
MediaEnemIDHM4.plot.scatter('NU_NOTA_REDACAO', 'IDHM', s=6, ax = ax[1,1], xlim = (maxRedacao, mean_REDACAO), ylim =(minIDHM, mean_IDHM))
```
